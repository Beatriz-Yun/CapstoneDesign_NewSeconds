```python

# 파일에 저장(DB에 저장)
def saveNewsToFile(titleList,contentsList,imageList):
    f = open('result.txt', 'w', encoding='utf-8')
    for i in range(len(titleList)):
        f.write(titleList[i] + "\n")
        f.writelines(contentsList[i]+"\n")
        f.writelines(imageList[i])
        f.writelines("\n\n")
    f.close()


# 더보기
def showMore():
        try:
            while True:
                driver.find_element_by_xpath('//*[@id="main_content"]/div/div[2]/div[2]/div/a').click()
        except exceptions.ElementNotVisibleException as e:
            pass
        except Exception as e:
            pass
# 크롤링
def subCrawl(soup):
    titleList=list()
    contentsList=list()
    imageList=list()
    articles = soup.find_all('div', {'class': 'cluster_group _cluster_content'})
    for i in range(len(articles)): 
        images=[]
        imagesURL=[]

        temp = articles[i].find_all('div', {'class': 'cluster_text'})[0]
        print(temp.a.text)

        conURL = temp.a.get('href')
        html2 = session.get(conURL,headers=headers).content
        soup2 = BeautifulSoup(html2, 'html.parser')
        contents = soup2.find('div', id= "articleBodyContents").text.replace('\n','')
        title=soup2.find('h3',id="articleTitle").text
        
        images=soup2.find_all('span', {'class','end_photo_org'})

        for i in range(len(images)):
            imagesURL.append(images[i].find("img")["src"])
            print(imagesURL[i])
        titleList.append(title)
        contentsList.append(contents)
        imageList.append(imagesURL)
    saveNewsToFile(titleList,contentsList,imageList)
        
# 페이징
def mainCrawl():    
    # 정치=100 경제=101 사회=102 생활/문화=103 세계=104 IT/과학=105
    for i in range(100, 102):
        main_url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1="+str(i)
        driver.get(main_url)
        # '헤드라인 더보기' 버튼이 있다면 누르기       
        showMore()
        driver.implicitly_wait(3)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')        
        subCrawl(soup)
    driver.quit()


def summarizeTitle(originTitle):
    okt = Okt()
    ko_doc_noun = okt.nouns(originTitle) # okt.nouns를 이용해 명사만 추출
    #print(ko_doc_noun)
    resultTitle=''
    for i in ko_doc_noun:
        resultTitle=resultTitle+' '+i
    return resultTitle

    
    
def searchImage(resultTitle):
    pass
```
